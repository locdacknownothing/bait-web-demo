```python
import pickle
import matplotlib.pyplot as plt

history_path = (
    # "/mnt/data/src/bait/train/sim/history/mobilenet_v3_large/history_seg_30k_margin1.0.pkl"
    "/mnt/data/src/bait/train/sim/history/resnet50/history_seg_30k_margin1.0.pkl"
)

with open(history_path, "rb") as f:
    history = pickle.load(f)

train_loss_history = history["train_loss"]
val_loss_history = history["val_loss"]
```

```python
history.keys()
```

```python
last_epoch = len(train_loss_history)
print(last_epoch)
```

```python
# Plot the loss history
plt.plot(train_loss_history, label="Training Loss")
plt.plot(val_loss_history, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
import numpy as np
np_val_loss = np.array(val_loss_history)
best_epoch = np.argmin(np_val_loss) + 1
print("Best epoch:", best_epoch)
```

```python
epoch = best_epoch

print(train_loss_history[epoch - 1])
print(val_loss_history[epoch - 1])
```
