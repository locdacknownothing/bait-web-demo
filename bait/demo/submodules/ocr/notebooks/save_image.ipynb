```python
import pickle 

def load_object(load_file: str) -> object:
    with open(load_file, "rb") as file:
        res = pickle.load(file)

    return res 
```

```python
output_ocr = load_object("../data/test_string_dict.pkl") 
len(output_ocr)
```
